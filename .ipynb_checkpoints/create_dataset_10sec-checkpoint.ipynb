```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
genres = ["blues", "classical", "country", "disco", "hiphop", "jazz", "metal", "pop", "reggae", "rock"]
```

```python
dict_ = {"blues" : [],
         "classical" : [],
         "country" : [],
         "disco" : [],
         "hiphop" : [],
         "jazz" : [],
         "metal" : [],
         "pop" : [],
         "reggae" : [],
         "rock" : []
        }
```

```python
dict_ = {"blues" : [],
         "classical" : [],
         "country" : [],
         "disco" : [],
         "hiphop" : [],
         "jazz" : [],
         "metal" : [],
         "pop" : [],
         "reggae" : [],
         "rock" : []
        }


for i in range(len(genres)):
    for j in range(100):
        if j < 10:
            scale_file = "data/genres_original/" + genres[i] + "/" + genres[i] + ".0000" + str(j) + ".wav"
        else:
            scale_file = "data/genres_original/" + genres[i] + "/" + genres[i] + ".000" + str(j) + ".wav"
        try:
            scale, sr = librosa.load(scale_file)
        except:
            print(scale_file)
        hop = int(scale.shape[0]/299)
        mel_spectrogram = librosa.feature.melspectrogram(y=scale, sr=sr, n_fft=2048, hop_length=hop, n_mels=128, window="hann")
        log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
        dict_[genres[i]].append(log_mel_spectrogram)
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(log_mel_spectrogram, 
                         x_axis="time",
                         y_axis="mel", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
X_list = []
y_list = []
for i, genre in enumerate(genres):
    for j in range(len(dict_[genre])):
        X_list.append(dict_[genre][j])
        y_list.append(genre)
        
print(X_list)
print(y_list)
```
